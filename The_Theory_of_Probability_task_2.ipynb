```python
import numpy as np
from math import factorial
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
# Задачи 1 и 2
# Контрольная работа состоит из пяти вопросов. На каждый вопрос приведено четыре варианта ответа, один из которых правильный. Случайная величина X задаёт число правильных ответов при простом угадывании. Найдите математическое ожидание данной случайной величины. Найдите дисперсию случайной величины X.
# Подсказка: постройте закон распределения случайной величины X.
# Solution: X - the number of right options. The distribution of this random variable is a the variable value compliance with taken probabilities. Let's take Bernoulli's formula for separate discrete random variable probability calculation:
```

```python
# p - any of five questions correct answer probability;
p = 1/4 
#  one need to answer correctly one of four questions;
p
```

```python
# n - questions quantity
# k - correct answers quantity
X = np.arange(0, 6) #[0, 1, 2, 3, 4, 5] = the number of correct answers
P = [] # relevant discrete random variable probability
n = 5
sum = 0
for k in range(0, 6):
    
    Pk = combinations(n, k) * (p**k) * ((1-p)**(n-k))
    P.append(Pk)
    sum = sum + Pk
    print(X[k], Pk)
print('Total quantity of probabilities:', sum)
```

```python
Mx = X.dot(P)
Mx
```

```python
y_values = X - Mx
print(y_values)
```

```python
z_values = y_values**2
Dx = z_values.dot(P)
print(Dx)
```

```python
# Задача 3
# Пользователь получает в среднем 10 писем со спамом на свой почтовый ящик за сутки. Найти число N, такое, что с вероятностью 0.95 пользователь получит не более N писем со спамом за текущий день.
# Solution: Random variable X takes values 0, 1, 2,... and relevant probabilities are expressed with the Poisson formula
```

```python
def poisson_proba(k: int, lambda_: float) -> float:
    """The Poisson Formula.
    """
    
    return (lambda_** k) * (np.exp(-lambda_)) / np.math.factorial(k)
```

```python
lambda_ = 10
sum = 0
i = -1
while (sum <= 0.95):
    i = i+1
    sum = sum + poisson_proba(k=i, lambda_=lambda_)
print('The answer: N=', i)
```
