```python
## Имеется колода в 52 карты. Найти число возможностей вытянуть из неё 4 карты так, чтобы среди них был хотя бы один туз.
import numpy as np
```

```python
## Для того, чтобы найти искомое число, нужно посчитать число исходов, соответствующих вытягиванию  1 ,  2 ,  3  и  4  тузов, а затем сложить эти числа. Все эти значения считаются аналогично. Например, посчитаем число исходов в первом случае. Для этого нужно посчитать число возможностей вытащить один туз из «колоды» в  4  туза ( C14 ) и число возможностей вытащить  3  произвольные карты из оставшейся колоды в  48  карт ( C348 ):
N1=C14⋅C348. 

Аналогично для  i≤4 :
Ni=Ci4⋅C4−i48.
def combinations(n: int, k: int) -> int:
    
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
```

```python
## Другой способ посчитать то же самое значение — посчитать общее число исходов ( C452 ) и вычесть из него число исходов, при которых в руке вообще не оказалось туза ( C448 ):
sum(combinations(4, i) * combinations(48, 4 - i) for i in range(1, 5))
```

```python
## Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Случайным образом студент вытягивает 3 вопроса. Какова вероятность, что все выбранные вопросы знакомы студенту? Какова вероятность что два из трёх вопросов знакомы студенту?
# Общее число исходов при вытягивании  3  билетов из  60  равно  C360=60!3!⋅57!=60⋅59⋅586 .
# Число исходов, благоприятных для события «все билеты знакомые», равно  C350=50⋅49⋅486 . Вероятность первого события:
```

```python
combinations(50, 3) / combinations(60, 3)
```

```python
## Чтобы посчитить число благоприятных исходов для второго события, надо посчитать число возможностей выбрать  2  знакомых билета (из  50 ) и  1  незнакомый билет (из  10 ). Число таких исходов:  C250⋅C110=50⋅492⋅10 . Вероятность второго события:
combinations(50, 2) * combinations(10, 1) / combinations(60, 3)
```

```python
## Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.
# Вопрос: какова вероятность, что вы действительно больны ей?
# Подсказка: используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

# Solution: Событие  A  — больны этой болезнью, событие  B  — тест на эту болезнь оказался положительным.

# В этом случае нам нужно найти вероятность  P(A|B)  (поскольку тест дал положительный результат). По формуле Байеса:
# P(A|B)=P(B|A)⋅P(A)P(B) 

# Раскроем знаменатель по формуле полной вероятности следующим образом:
# P(B)=P(B|A)⋅P(A)+P(B|A¯¯¯¯)⋅P(A¯¯¯¯) 

# Подстановка известных величин:
# P(B)=0.99 * 0.001 + 0.01 * 0.999 = 0.01098
```

```python
P_d = 0.001 * 0.99 + 0.999 * 0.01

print(P_d)
```

```python
# Подставляя это значение в формулу Байеса, получаем:
# P(A|B)=0.99 * 0.001 / 0.01098 = 0.0902
```

```python
0.001 * 0.99 / P_d
```
